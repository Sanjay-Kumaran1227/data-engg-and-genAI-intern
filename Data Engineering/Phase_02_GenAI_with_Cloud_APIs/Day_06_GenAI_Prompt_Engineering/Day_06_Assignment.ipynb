```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_zero_shot = ChatGroq(model="llama-3.3-70b-versatile", temperature=0.2)

zero_shot_prompt = PromptTemplate(
    input_variables=["instruction"],
    template="""Follow the instruction and provide the output clearly:

Instruction:{instruction}"""
)

zero_shot_parser = StrOutputParser()

zero_shot_chain = zero_shot_prompt | llm_zero_shot | zero_shot_parser

instruction_zero_shot = "What is Big Data Analytics"
response_zero_shot = zero_shot_chain.invoke({
    "instruction": instruction_zero_shot
})

print("\nZero-Shot Output:\n", response_zero_shot)
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_few_shot = ChatGroq(model="llama-3.3-70b-versatile", temperature=0.2)

few_shot_prompt = PromptTemplate(
    input_variables=["text"],
    template="""Classify the sentiment of the following text as 'Positive', 'Negative', or 'Neutral'.

Examples:
Text: I love this movie! It's fantastic.
Sentiment: Positive

Text: The service was terrible and slow.
Sentiment: Negative

Text: This product is okay, nothing special.
Sentiment: Neutral

Text: {text}
Sentiment:"""
)

few_shot_parser = StrOutputParser()

few_shot_chain = few_shot_prompt | llm_few_shot | few_shot_parser

text_to_classify = "The food was delicious, but the wait time was too long."
response_few_shot = few_shot_chain.invoke({
    "text": text_to_classify
})

print("\nFew-Shot Output (Sentiment Classification):\n", response_few_shot)
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_cot = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

cot_prompt = PromptTemplate(
    input_variables=["question"],
    template="""Solve the following problem. Think step-by-step and show your reasoning before providing the final answer.

Question: {question}
"""
)

cot_parser = StrOutputParser()

cot_chain = cot_prompt | llm_cot | cot_parser

question_cot = "If a car travels at 60 miles per hour for 3 hours, how far does it travel?"
response_cot = cot_chain.invoke({
    "question": question_cot
})

print("\nChain of Thought Output:\n", response_cot)
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_role = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

role_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a witty and sarcastic comedian. Respond to all prompts with humor and a touch of sarcasm."),
    ("user", "{question}")
])

role_parser = StrOutputParser()

role_chain = role_prompt | llm_role | role_parser

question_role = "Tell me about the weather today."
response_role = role_chain.invoke({
    "question": question_role
})

print("\nRole Prompting Output:\n", response_role)
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_tot = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

tot_prompt = PromptTemplate(
    input_variables=["problem"],
    template="""For the following problem, brainstorm at least three distinct approaches to solve it. For each approach, describe it briefly and explain its pros and cons.

Problem: {problem}
"""
)

tot_parser = StrOutputParser()

tot_chain = tot_prompt | llm_tot | tot_parser

problem_tot = "How can a new cafe attract its first 100 customers in a competitive market?"
response_tot = tot_chain.invoke({
    "problem": problem_tot
})

print("\nTree of Thought (Simplified) Output:\n", response_tot)
```
